```python
import pandas as pd
```

```python
#To read the file
kai = pd.read_csv('Board2.csv')
kai
```

```python
#Creating the file "Vis" for a condition that the only data will 
    #show are students with a hometown of Visayas and passed math class

#Trying the example from the Laboratory manual.
```

```python
#Filter the rows
Vis = kai[(kai['Hometown'] == 'Visayas') & (kai['Math']<70)]

#Select needed columns only, note that 2 brackets are needed for multiple columns
Vis = Vis[['Name', 'Gender', 'Track', 'Math']]
Vis

#The output is the same with the example. therefore the format is correct.
```

```python
#doing condition 'a'
#Filter again, the student must have instrumentation track, Hometown is luzon, 
                                                        #and passed Electronics
Instru = kai[(kai['Track'] == 'Instrumentation') & 
            (kai['Hometown'] == 'Luzon') & 
            (kai['Electronics']>70)]

#Select the neeeded Columns again
Instru = Instru[['Name', 'GEAS', 'Electronics']]
Instru
```

```python
#Remember that the courses/classes are: Math, Electronics, GEAS, and Communication
```

```python
#doing condition 'b'
#Filter again LOL, the student must be Female, constant hometown is Mindanao 
                                                                #with ave = 55
kai['Average'] = (kai['Math'] + kai['Electronics'] + kai['GEAS'] + kai['Communication']) / 4
Mindy = kai[(kai['Gender'] == 'Female') &
            (kai['Hometown'] == 'Mindanao') &
            (kai['Average'] >= 55)]
Mindy = Mindy[['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

```python
#Probelem 2
```

```python
import matplotlib.pyplot as plt
```

```python
#Gender case

avg_by_gender = kai.groupby('Gender')['Average'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_by_gender.index, avg_by_gender.values)
plt.title("Average Grade by Gender")
plt.ylabel("Average Grade")
plt.show()
```

```python
#Track case

avg_by_track = kai.groupby('Track')['Average'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_by_track.index, avg_by_track.values)
plt.title("Average Grade by Track")
plt.ylabel("Average Grade")
plt.show()
```

```python
#Hometown case

avg_by_hometown = kai.groupby('Hometown')['Average'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_by_hometown.index, avg_by_hometown.values)
plt.title("Average Grade by Hometown")
plt.ylabel("Average Grade")
plt.show()
```
